```python
names=[]
desc=[]
reviews=[]
price=[]

model=soup.find_all("div",class_="_4rR01T")
for i in model:
    x=i.text
    names.append(x)
   




```

```python
print(names)
```

```python
prices=soup.find_all("div",class_="_30jeq3 _1_WHN1")
for i in prices:
    x=i.text
    price.append(x)
```

```python
price
```

```python
des=soup.find_all("ul",class_="_1xgFaf")
for i in des:
    x=i.text
    desc.append(x)
```

```python
desc
```

```python
rev=soup.find_all("span",class_="_2_R_DZ")
for i in rev:
    x=i.text
    reviews.append(x)
```

```python
reviews
```

```python
df=pd.DataFrame({"Model_Name":names,"Model_Price":price,"Model_desc":desc,"Model_review":reviews})
```
