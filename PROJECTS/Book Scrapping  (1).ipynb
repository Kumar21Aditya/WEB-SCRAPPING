```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url="https://books.toscrape.com/catalogue/page-1.html"
r=requests.get(url)
print(r)
```

```python
soup=bs(r.text,"lxml")
print(soup)
```

```python
name=soup.find_all("h3")
```

```python
name
```

```python
NAME=[]
for i in name:
    x=i.text
    NAME.append(x)
```

```python
NAME
```

```python
price=soup.find_all("p",class_="price_color")
price
```

```python
PRICE=[]
for i in price:
    x=i.text
    PRICE.append(x)
```

```python
PRICE
```

```python
import re
```

```python
rating=soup.find_all("p",class_=re.compile("star-rating"))
rating
```

```python
RATING=[]
for i in soup.find_all("p",class_=re.compile("star-rating")):
    x=i['class'][1]
    RATING.append(x)
print(RATING)    
```

```python
availability=soup.find_all("p",class_="instock availability")
availability
```

```python
AVAILABILITY=[]
for i in availability:
    x=i.text.strip()
    AVAILABILITY.append(x)
print(AVAILABILITY)    
    
```

```python
len(AVAILABILITY)
```

# For all pages

```python
NAME=[]
RATING=[]
PRICE=[]
STOCK=[]
```

```python
for i in range(1,51):
    url="https://books.toscrape.com/catalogue/page-"+str(i)+".html"
    print(url)
    r=requests.get(url)
    print(r)
    
    soup=bs(r.text,"lxml")
    
    name=soup.find_all("h3")
    for i in name:
        x=i.text
        NAME.append(x)
    print(len(NAME))  
    
    price=soup.find_all("p",class_="price_color")
    for i in price:
        x=i.text
        PRICE.append(x)
    print(len(PRICE))
    
    stock=soup.find_all("p",class_="instock availability")
    for i in stock:
        x=i.text.strip()
        STOCK.append(x)
    print(len(STOCK))
    
    for i in soup.find_all("p",class_=re.compile("star-rating")):
        x=i["class"][1].strip()
        RATING.append(x)
    print(len(RATING))    
    
    
    
```

```python
df2=pd.DataFrame({"BOOK-NAME":NAME,"BOOK-PRICE":PRICE,"RATING":RATING,"CURRENT STATUS":STOCK})
```
